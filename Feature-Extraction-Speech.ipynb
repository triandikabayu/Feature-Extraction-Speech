```python
#Zero Crossing Rate
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display


x, sr = librosa.load('OAF_base_happy.wav')

ipd.Audio(x, rate=sr)

# Visualisasi audio
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)

# Select portion
n0 = 6500
n1 = 7500
plt.figure(figsize=(14, 5))
plt.plot(x[n0:n1])

# Compute zero-crossings for the selected audio portion
zero_crossings = librosa.zero_crossings(x[n0:n1], pad=False)
print(sum(zero_crossings))
# Compute zero-crossing rate for the entire audio signal
zcrs = librosa.feature.zero_crossing_rate(x)
print(zcrs.shape)
# Visualize zero-crossing rate
plt.figure(figsize=(14, 5))
plt.plot(zcrs[0])
```

```python
#CHROMAGRAM
import librosa
import librosa.display
import matplotlib.pyplot as plt

# load audio file
audio_file = "OAF_base_happy.wav"
y, sr = librosa.load(audio_file)

# extract chromagram feature
chromagram = librosa.feature.chroma_stft(y=y, sr=sr)

# plot chromagram
plt.figure(figsize=(10, 4))
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma',
cmap='coolwarm')
plt.colorbar()
plt.title('Chromagram')
plt.tight_layout()
plt.show()
```

```python
#SPECTROGRAM
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Load audio file
audio_file = 'OAF_base_happy.wav'
y, sr = librosa.load(audio_file, sr=None)

# Extract spectrogram features
spec = np.abs(librosa.stft(y))

# Visualize spectrogram features
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(spec, ref=np.max),
y_axis='log', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```

```python
#RMSE
import librosa
import librosa.display
import matplotlib.pyplot as plt

# load audio file
y, sr = librosa.load('OAF_base_happy.wav')

# extract RMS energy feature
rmse=librosa.feature.rms(y=y)[0]
print(rmse.shape)

# get the times corresponding to each feature frame
times = librosa.times_like(rmse)

# plot the RMSE feature over time
plt.semilogy(times, rmse, label='RMS Energy')
plt.show()
```

```python
#MFCC
import sklearn
import librosa, librosa.display
import matplotlib.pyplot as plt

#load audio
x, fs = librosa.load('OAF_base_happy.wav')
librosa.display.waveshow(x, sr=fs)

# MFCC Processing
mfccs = librosa.feature.mfcc(x, sr=fs)
mfcss = sklearn.preprocessing.scale(mfccs, axis=1)
librosa.display.specshow(mfccs, sr=fs, x_axis='time')
```

```python
#SPECTRAL CENTROID
import numpy as np
import scipy
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa, librosa.display

# load audio file
x, sr = librosa.load('OAF_base_happy.wav')

# extract spectral centroid feature
cent = librosa.feature.spectral_centroid(y=x, sr=sr)

# visualize spectral centroid feature
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogy(cent.T, label='Spectral centroid')
plt.ylabel('Hz')
plt.xticks([])
plt.xlim([0, cent.shape[-1]])
plt.legend()
plt.show()
```
